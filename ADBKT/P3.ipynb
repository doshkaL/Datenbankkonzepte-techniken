```python


# Verbindung zur Cassandra-Datenbank herstellen
from cassandra.cluster import Cluster

import uuid 

cluster = Cluster(['172.19.0.5'], port=9042)
session = cluster.connect()  
session.set_keyspace('tourdelight')

#
#TRUNCATE TABLE profil;

# esttelung von KEYSPACE

#sql1 ="CREATE KEYSPACE tourdelight WITH replication = {'class': 'SimpleStrategy', 'replication_factor': 3};"

#use tourdelight; SELECT * FROM profil;
# erstellen von  tourdelight.profil
#sql2 = """
#CREATE TABLE tourdelight.profil (
 # id uuid PRIMARY KEY, 
  #username text,
  #email text,
  #name text,
  #vorname text,
  #geburtsdatum date,
  #geschlecht text
#);
#"""


# Profil anlegen
def create_profile(username, email, name, vorname, geburtsdatum, geschlecht):
  id = uuid.uuid4()
  query = """
    INSERT INTO profil (id, username, email, name, vorname, geburtsdatum, geschlecht)
    VALUES (%s, %s, %s, %s, %s, %s, %s)
  """
  values = (id, username, email, name, vorname, geburtsdatum, geschlecht)
  session.execute(query, values)
  return id  # Return the generated UUID

# Profil löschen
def delete_profile(id):
  query = """
    DELETE FROM profil
    WHERE id = %s
  """
  values = (id,)
  session.execute(query, values)

# Profil auslesen
def read_profile(id):

  query = """
    SELECT *
    FROM profil
    WHERE id = %s
  """
  values = (id,)
  result = session.execute(query, values)
  # Rückgabe des ersten Eintrags im Ergebnis-Set, oder `None`, wenn kein Eintrag vorhanden ist
  return result.one() if result else None

# Profil anlegen and get the generated UUID
profile_id = create_profile("user", "test@example.com", "Mustermann", "Max", "1980-01-01", "männlich")
profile_id2 = create_profile("user2", "test@example2.com", "jawasser", "ffark", "1988-01-01", "männlich")

# Profil auslesen
profile = read_profile(profile_id)
if profile:
    print(profile)
else:
    print("Profile not found.")
    
    # Profil2 auslesen
profile2 = read_profile(profile_id2)
if profile2:
    print(profile2)
else:
    print("Profile not found.")

# Delete the profile
#delete_profile(profile_id)

# Print a message to confirm that the profile has been deleted
#print("Profile deleted successfully.")
```
